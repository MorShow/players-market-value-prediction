```python
import pandas as pd
import numpy as np
from functools import reduce

import os

%matplotlib inline
```

```python
tables = []
names = []

for name in os.listdir(r'data'):
    if name != 'train_stats.csv' and name != 'test_stats.csv':
        tables.append(pd.read_csv('data\\' + name))
        names.append(name)
```

```python
league_tables = []
league_names = []

for i in range(1, 6):
    league = tables[(i-1)*4:i*4]
    
    df1 = league[0].merge(league[1], on=['Name', 'Team', 'Age', 'Position', 'Apps', 'Mins', 'Rating'])
    df2 = df1.merge(league[2], on=['Name', 'Team', 'Age', 'Position', 'Apps',
                                   'Mins', 'Rating', 'Goals', 'Assists', 'SpG'])
    df3 = df2.merge(league[3], on=['Name', 'Team', 'Age', 'Position', 'Apps',
                                   'Mins', 'Rating', 'Assists', 'KeyP', 'PS'])
    
    league_tables.append(df3)
    league_names.append(names[(i-1)*4].split('_')[0])
```

```python
for tab, name in zip(league_tables, league_names):
    print(f'League: {name}')
    print(f'Shape: {tab.shape}')
    display(tab.sample(1))
```

```python
table = pd.concat(league_tables)
```

```python
table.reset_index(inplace=True, drop=True)
```

```python
indices = np.arange(table.shape[0])
np.random.shuffle(indices)

train_size = int(0.8 * table.shape[0])

table.iloc[indices[:train_size]].to_csv('data/train_stats.csv', index=False)
table.iloc[indices[train_size:]].to_csv('data/test_stats.csv', index=False)
```
